```python
en = '''[Rachel Green (Jennifer Aniston, жен, 44, 1950)] Oh/ are you sure you're ok?
[Ross Geller (David Schwimmer, муж, 32, 1962)] Yeah.
[Rachel Green (Jennifer Aniston, жен, 44, 1950)] Does it still hurt?
[Ross Geller (David Schwimmer, муж, 32, 1962)] Yeah.
[Phoebe Buffay (Lisa Kudrow, жен, 54, 1940)] What a neat idea. All your clothes match. I'm gonna do this.'''
```

```python
ru = '''[Рейчел Грин (Ольга Голованова, жен, 42, 1952)] Тебе лучше?
[Росс Геллер (Александр Рахленко, муж, 39, 1955)] Да.
[Рейчел Грин (Ольга Голованова, жен, 42, 1952)] Очень болит?
[Росс Геллер (Александр Рахленко, муж, 39, 1955)] Да.
[Фиби Буффе (Марина Дюжева, жен, 38, 1956)] Потрясающе/ всё в одной гамме. Я тоже так сделаю.'''
```

```python
import re
```

```python
en_split = re.split(r'\n?\[(.+?)\] ', en)[1:]
ru_split = re.split(r'\n?\[(.+?)\] ', ru)[1:]
```

```python
len(en_split) == len(ru_split)
```

```python
match_meld = pd.DataFrame({'en_actor': en_split[::2], 
                           'ru_actor': ru_split[::2], 
                           'text_en': en_split[1::2], 
                           'text_ru': ru_split[1::2]})
```

```python
match_meld
```

Частый знак в русской транкрибации `/` - подъем тона, здесь: крайне проблемен при выравнивании текстов :с

```python
match_meld['preprocessed'] = match_meld.text_en.apply(lambda x: ' '.join(re.findall(r'\w+', x.lower())))
```

```python
match_meld
```

```python
imp_part['preprocessed'] = imp_part.Utterance.apply(lambda x: ' '.join(re.findall(r'\w+', x.lower())))
```

```python
match_meld = match_meld.merge(imp_part[['preprocessed', 'Sentiment', 'Emotion']], on='preprocessed')
```

```python
match_meld.drop('preprocessed', axis=1, inplace=True)
```
